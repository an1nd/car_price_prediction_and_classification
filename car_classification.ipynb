```python
# Importing all libraries :

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv('cars_class.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

DATA PREPROESSING

```python
#checking any duplicates
df.duplicated(subset = ['ID']).sum()
```

```python
df = df.drop(['ID'], axis =1) # Index column is an insignificant column.
```

```python
df.isnull().sum()
```

DATA VISUALIZATION

```python
cars_numeric = df.select_dtypes(include =['int64'])
cars_numeric
```

```python
# plt.figure(figsize = (20,20))
# sns.heatmap(df.corr(), annot = True ,cmap = 'YlGnBu')
# plt.show()
```

```python
df_train, df_test = train_test_split(df, train_size = 0.8, test_size = 0.2, random_state = 100)
```

```python
df_train
```

```python
df_test
```

```python
y_train = df_train.pop('Class')
X_train = df_train
```

```python
y_test = df_test.pop('Class')
X_test = df_test
```

DECISION TREE CLASSIFIER

```python
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train)
dtree_predictions = dtree_model.predict(X_test)
```

```python
# creating a confusion matrix
cm = confusion_matrix(y_test, dtree_predictions)
cm
```

```python
dtree_accuracy = accuracy_score(y_test,dtree_predictions)
dtree_accuracy
```

```python
dtree_f1 = f1_score(y_test,dtree_predictions,average='micro')
dtree_f1
```

SVM (SUPPORT VECTOR MACHINE) CLASSIFIER

```python
from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(X_test)
```

```python
# model accuracy for X_test  
accuracy_svm = svm_model_linear.score(X_test, y_test)
accuracy_svm
```

```python
# creating a confusion matrix
cm_svm = confusion_matrix(y_test, svm_predictions)
cm_svm
```

```python
svm_f1 = f1_score(y_test,svm_predictions,average='micro')
svm_f1
```

KNN (K-NEAREST NEIGHBORS CLASSIFIER)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)
knn_predictions = knn.predict(X_test)
```

```python
knn_accuracy = knn.score(X_test, y_test)
knn_accuracy
```

```python
knn_cm = confusion_matrix(y_test, knn_predictions)
knn_cm
```

```python
knn_f1 = f1_score(y_test,knn_predictions,average='micro')
knn_f1
```

NAIVE-BAYES CLASSFIER

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)
```

```python
# accuracy on X_test
gnb_accuracy = gnb.score(X_test, y_test)
gnb_accuracy
```

```python
# creating a confusion matrix
gnb_cm = confusion_matrix(y_test, gnb_predictions)
gnb_cm
```

```python
gnb_f1 = f1_score(y_test,gnb_predictions,average='micro')
gnb_f1
```

```python
# SVM HAS THE HIGHEST ACCURACY WHEN COMPARED WITH DECISION TREE, KNN, NAIVE-BAYES CLASSIFIERS
```
